# Data preprocessing

## We can see here that we need to remove the user names,punctuations, emojis , Numbers and \n

```python
df_copy = df.copy()
```

```python
def first_preprocessing(text):

    text = re.sub(r'@\w+', '', text)            # usernames
    text = re.sub(r'\d+', '', text)             # numbers
    text = text.replace('\n', ' ')              # \n
    return text
```

```python
df_copy['text'] = df_copy['text'].apply(first_preprocessing)
```

```python
df_copy.tail(10)
```

# removing emojis

```python
emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  
        u"\U0001F300-\U0001F5FF"  
        u"\U0001F680-\U0001F6FF"  
        u"\U0001F1E0-\U0001F1FF"  
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
```

```python
def removing_emojis(text):
  text = emoji_pattern.sub(r'', text)
  return text
```

```python
df_copy['text'] = df_copy['text'].apply(removing_emojis)
```

```python
df_copy.head(10)
```

# remove url

```python
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+')
    return url_pattern.sub(r'', text)
```

```python
df_copy['text'] = df_copy['text'].apply(remove_urls)
```

```python
df_copy.head()
```

# removing stopwords

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

# define the Arabic stop words list
ar_stopwords = set(stopwords.words('arabic'))
```

```python
def removing_stopwords(text):
    words = text.split()

    filtered_words = [word for word in words if word not in ar_stopwords]

    text = ' '.join(filtered_words)

    return text
```

```python
df_copy['text'] = df_copy['text'].apply(removing_stopwords)
```

```python
df_copy.head(10)
```

# remove elongation and extra whitspace

```python
def remove_elongation(text):
    text = re.sub(r'(\w)\1{2,}', r'\1\1', text)    # remove text elongation
    text = re.sub(r'\s+', ' ', text)          # remove extra whitespaces
    return text.strip()
```

```python
df_copy['text'] = df_copy['text'].apply(remove_elongation)
```

```python
df_copy.head(10)
```

# remove punctuation

```python
import string
def remove_punctuation(text):
  text = ''.join([c for c in text if c not in string.punctuation])
  text = re.sub(r'ุ', ' ', text) 
  return text.strip()
```

```python
df_copy['text'] = df_copy['text'].apply(remove_punctuation)
```

```python
df_copy.head(10)
```

# character normalization

```python
def character_normalization(text):
  preprocessed_text = re.sub("[ุฅุฃุขุง]", "ุง", text)
  preprocessed_text = re.sub("ู", "ู", preprocessed_text)
  preprocessed_text = re.sub("ุค", "ุก", preprocessed_text)
  preprocessed_text = re.sub("ุฆ", "ุก", preprocessed_text)
  preprocessed_text = re.sub("ุฉ", "ู", preprocessed_text)
  preprocessed_text = re.sub("ฺฏ", "ู", preprocessed_text)
  preprocessed_text = re.sub("ฺค", "ู", preprocessed_text)
  preprocessed_text = re.sub("ฺ", "ุฌ", preprocessed_text)
  preprocessed_text = re.sub("ฺ", "ุฒ", preprocessed_text)
  preprocessed_text = re.sub("ูพ", "ุจ", preprocessed_text)
  return preprocessed_text
```

```python
df_copy['text'] = df_copy['text'].apply(character_normalization)
```

```python
df_copy.head(10)
```

# remove diacritics

```python
arabic_diacritics = re.compile("""
                             ู    | # Tashdid
                             ู    | # Fatha
                             ู    | # Tanwin Fath
                             ู    | # Damma
                             ู    | # Tanwin Damm
                             ู    | # Kasra
                             ู    | # Tanwin Kasr
                             ู    | # Sukun
                            ู    | # Tatwil/Kashida
                         """, re.VERBOSE)
```

```python
def remove_diacritics(text):
  preprocessed_text = re.sub(arabic_diacritics, '', text)
  return preprocessed_text
```

```python
df_copy['text'] = df_copy['text'].apply(remove_diacritics)
```

```python
df_copy.head(10)
```

# remove english

```python
def remove_english(text):
    english_pattern = re.compile("[a-zA-Z]+")
    return english_pattern.sub(r'', text)
```

```python
df_copy['text'] = df_copy['text'].apply(remove_english)
```
