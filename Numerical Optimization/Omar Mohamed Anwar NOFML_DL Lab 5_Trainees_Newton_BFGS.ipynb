## Newton's Method Optimization
### Single Variable
- Implement the <b>Gradient decent and Newton's method</b> as a function and find the minimum of the following function:
    - $𝒇(𝒙)=𝒙^𝟐−𝟏𝟓𝒙+𝟏$,        
- Draw the functions and check out the minimum.
- Draw the 1st derivative of the function and check out the root.
- Draw 2nd derivative of the function and check out the sign at the minimum.
- The stop condition is grad < epsilon.
- Try different epsilons and initial conditions.
- Compare the results with Gradient descent.
##### The function should return the solution and the gradient.

##### Notes:
- You will need to pass the expression of the function, first, and second derivatives to the optimizer.
- You can use lambda expression to construct the function, first, and second derivatives.

```python
x = np.linspace(-50,62,1000)
x
```

```python
f = lambda x: x**2 - 15*x +1
Df = lambda x: 2*x -15
DDf = lambda x: 2
```

```python
y = f(x)
dy = Df(x)
ddy = DDf(x)
```

```python
# Gradient descent with learning rate = 0.01, x0=40, epsilon = 0.1
x0 = 40
iterations = 0
alpha = 0.01
epsilon = 0.1
gradient = 1

while True : 
  gradient  = Df(x0)
  if gradient > epsilon :
    x0 = x0 - alpha*gradient
    iterations += 1
  else : break
print('GD Results with x0 =40, lr=0.01, epsilon=0.1')

print(f'Found solution after {iterations} iterations.')
print("x_min = ",x0)
print("Gradient = ",gradient)
```

```python
# Newton's method without learning rate, x0=40, epsilon = 0.1
x0 = 40
iterations = 0
epsilon = 0.1
first_der = 1

while True:
  first_der  = Df(x0)
  if first_der > epsilon :
    second_der = DDf(x0)
    x0 = x0 - first_der/second_der
    iterations += 1
  else : break

print('Newton Results with x0 =40, epsilon=0.1')
print(f'Found solution after {iterations} iterations.')
print("x_min = ",x0)
print("Gradient = ",first_der)
```

### Multi Variable

```python
from sympy import *
from sympy.plotting import plot3d
```

$𝒇(𝒙)=𝒙^𝟐y$ 

```python
x,y = symbols('x y')
f = x**2 * y
plot3d(f,size=(10,7))
plt.show()
```

```python
f_1 = lambda x: x[0]**2  + x[1]**2
Df_1 = lambda X: np.array([2*X[0],2*X[1]])
DDf_1 = lambda X: np.array([[2,0],
                         [0,2]])
```

```python
f_2 = lambda x: x[0]**2 + x*[1] + x[1]**2
Df_2 = lambda X: np.array([2*X[0]+X[1],2*X[1]+X[0]])
DDf_2 = lambda X: np.array([[2,1],
                        [1,2]])
```

```python
# First function
# Multi-variable Gradient descent with learning rate = 0.1,
# x0=(10,10), epsilon = 0.001
x0 = [10,10]
iterations = 0
epsilon = 0.001
alpha = 0.1
gradient_norm = 1

while True : 
  gradient  = Df_1(x0)
  gradient_norm = np.linalg.norm(gradient)
  if gradient_norm > epsilon :
    x0 = x0 - alpha*gradient
    iterations += 1
  else : 
    break

print(f'Found solution after {iterations} iterations.')
print("x_min = ",x0)
print("Gradient = ",gradient)
```

```python
# First function
# Multi-variable Newton's method without learning rate, 
# x0=(10,10), epsilon = 0.001

x0 = [10,10]
iterations = 0
epsilon = 0.001
gradient_norm = 1

while True : 
  gradient  = Df_1(x0)
  gradient_norm = np.linalg.norm(gradient)
  if gradient_norm > epsilon :
    hessian_inverse = np.linalg.inv(DDf_1(x0))
    x0 = x0 - hessian_inverse@gradient
    iterations += 1
  else : 
    break

print(f'Found solution after {iterations} iterations.')
print("x_min = ",x0)
print("Gradient = ",gradient)
```

## BFGS
- Implement the <b>BFGS</b> as a function and find the minimum of the following function:
    - $𝒇(𝒙)=𝒙^𝟐+y^2$
    - $𝒇(𝒙)=𝒙^𝟐+xy+y^2$  
    - $𝒇(𝒙)=𝒙^𝟐y$ 
- The stop condition is grad < epsilon.
- Try different epsilons and initial conditions.
- Compare the results from Gradient descent and Newton's method.

```python
## First function
## BFGD with xk-1 = (10,10) and x0 = (9,9), epsilon = 0.001
xk = np.array([10,10]).reshape(-1,1)
x0 = np.array([9,9]).reshape(-1,1)
iterations = 0
epsilon = 0.001
gradient_norm = 1
I = np.identity(2)
beta = I

while True : 
  gradient  = Df_1(x0)

  if np.linalg.norm(gradient) > epsilon :
    print(f" Iteration {iterations+1} ".center(100,'#'))
    print("xn_1 : ",xk)
    print("xn :",x0)
    print("Dfxn_1 : ",Df_1(xk))
    print("Dfxn : ",Df_1(x0))

    y = Df_1(x0) - Df_1(xk)
    yT = np.transpose(y)

    _x = x0 - xk
    _xT = np.transpose(_x)

    a = I - np.matmul(_x,yT)/np.dot(yT,_x)
    b = I - np.matmul(y,_xT)/np.dot(yT,_x)
    c = np.matmul(_x,_xT)/np.dot(yT,_x)

    beta = (a@beta)@b+c
    
    xk = x0 
    x0 = x0 - beta@gradient
       
    iterations += 1

    print("Beta Inverse : ",beta)
  else : 
    break

print(f'Found solution after {iterations} iterations.')
print("x_min = ",x0)
print("Gradient = ",np.linalg.norm(gradient))
```
