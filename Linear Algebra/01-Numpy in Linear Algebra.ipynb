## Calculate in paper then confirm your results using numpy linalg.norm().

### Create two vectors in 2, 3, and 4 dimensions.
### Perform vector addition, subtraction, and dot product of these vectors.

```python
v_2d_1 = np.array([1,2]) ; print(v_2d_1)
v_2d_2 = np.array([3,4]) ; print(v_2d_2)
sum_2d = np.add(v_2d_1,v_2d_2) ; print(sum_2d)
sub_2d = np.add(v_2d_1,-v_2d_2) ; print(sub_2d)
dot_product_2d = np.dot(v_2d_1,v_2d_2) ; print(dot_product_2d,'\n')

v_3d_1 = np.array([1,2,3]) ; print(v_3d_1)
v_3d_2 = np.array([4,5,6]) ; print(v_3d_2)
sum_3d = np.add(v_3d_1,v_3d_2) ; print(sum_3d)
sub_3d = np.add(v_3d_1,-v_3d_2) ; print(sub_3d)
dot_product_3d = np.dot(v_3d_1,v_3d_2) ; print(dot_product_3d,'\n')

v_4d_1 = np.array([1,2,3,4]) ; print(v_4d_1)
v_4d_2 = np.array([5,6,7,8]) ; print(v_4d_2)
sum_4d = np.add(v_4d_1,v_4d_2) ; print(sum_4d)
sub_4d = np.add(v_4d_1,-v_4d_2) ; print(sub_4d)
dot_product_4d = np.dot(v_4d_1,v_4d_2) ; print(dot_product_4d)
```
